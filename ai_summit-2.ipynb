```python
abbrev_data = pd.read_excel("/content/ABBREV.xlsx")
abbrev_data
```

```python
final_abbrev_data= abbrev_data.drop(columns=['NDB_No','Shrt_Desc','Water_(g)', 'Energ_Kcal','Ash_(g)','Fiber_TD_(g)','Choline_Tot_ (mg)','Beta_Crypt_(µg)','Lycopene_(µg)','Lut+Zea_ (µg)','GmWt_1','GmWt_Desc1','GmWt_2','GmWt_Desc2','Refuse_Pct'])
```

```python
nutrients_data = pd.read_csv("/content/Daily_Nutrient.csv")
nutrients_data
```

```python
import cvxpy as cp
import numpy as np

n = 8790
np.random.seed(1)
A = final_abbrev_data.to_numpy().T
b = nutrients_data.to_numpy()
c = np.ones((8790))

# Define and solve the CVXPY problem.
x = cp.Variable(n)
prob = cp.Problem(cp.Minimize(c.T@x),
                 [A @ x <= b])
prob.solve()

# Print result.
print("\nThe optimal value is", prob.value)
print("A solution x is")
print(x.value)
print("A dual solution is")
print(prob.constraints[0].dual_value)
```
